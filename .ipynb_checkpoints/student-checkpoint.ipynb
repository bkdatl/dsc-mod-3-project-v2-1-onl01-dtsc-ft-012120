## Dataset

The dataset is comprised of customer data collected from a syrian based telecom company and their telecom users in the united states. 

state - US State customer account, listed in state abbreviation 

account length 	- Number of months account active 

area code 	- area code of the user

phone number 	- phone number of the user 

international plan 	- yes/no value for if they pay for international calling plan 

voice mail plan 	- yes/no value if they pay for voicemail plan 

number vmail messages 	- Number of voicemails that are active on plan 

total day minutes 	- number of minutes used per day 

total day calls 	- number of calls used per day 

total day charge 	- total charge for used minutes 

total eve calls 	- number of minutes used during evening billing hours 

total eve charge 	- total charge of evening time calls 

total night minutes - total number of night minutes used during night billing hours 

total night calls 	- total number of calls placed during night billing hours 

total night charge 	- total charge of night billing hours 

total intl minutes 	- total initial minutes on the account 

total intl calls 	- total calls on the account 

total intl charge 	- total initial charge on the account 

customer service calls 	- number of times the customer has contacted customer service 

churn - Boolean value: True/False of customer churn

Model choices: 

The following models were chosen for this project for a few reasons. In computational complexity and optimization the no free lunch theorem is a result that states that for certain types of mathematical problems, the computational cost of finding a solution, averaged over all problems in the class, is the same for any solution method. No solution therefore offers a "short cut". In computing, there are circumstances in which the outputs of all procedures solving a particular type of problem are statistically identical.

Thus multiple models are chosen to compare against eachother in order to dertermine the most effective or simmilarly effective models.

### XGBoost 

```python
#Split data into tests and sets
labels = df_cleaned['churn']
labels_removed_df = df_cleaned.drop('churn', axis=1, inplace=False)
scaler = StandardScaler()
scaled_df = scaler.fit_transform(labels_removed_df)
X_train, X_test, y_train, y_test = train_test_split(scaled_df, labels)
```

```python

clf = xgb.XGBClassifier()
clf.fit(X_train, y_train)
training_preds = clf.predict(X_train)
val_preds = clf.predict(X_test)
training_accuracy = accuracy_score(y_train, training_preds)
val_accuracy = accuracy_score(y_test, val_preds)

print("Training Score: {:.5}%".format(training_accuracy * 100))
print("Test Score: {:.5}%".format(val_accuracy * 100))
```

```python
param_grid = {
    "learning_rate": [0.1],
    'max_depth': [6],
    'min_child_weight': [10],
    'subsample': [ 0.7],
    'n_estimators': [5, 30, 100, 250],
}
```

```python


grid_clf = GridSearchCV(clf, param_grid, scoring='accuracy', cv=None, n_jobs=1)
grid_clf.fit(scaled_df, labels)

best_parameters = grid_clf.best_params_

print("Using paramater grid, the model was optimized with the following optimal parameters: ")
for param_name in sorted(best_parameters.keys()):
    print("%s: %r" % (param_name, best_parameters[param_name]))

training_preds = grid_clf.predict(X_train)
val_preds = grid_clf.predict(X_test)
training_accuracy = accuracy_score(y_train, training_preds)
val_accuracy = accuracy_score(y_test, val_preds)

print("")
print("Training Score: {:.4}%".format(training_accuracy * 100))
print("Test Score: {:.4}%".format(val_accuracy * 100))

```
